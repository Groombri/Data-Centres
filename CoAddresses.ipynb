```python
# Takes a postcode (of data centre) and returns all business-owned properties
# in the area
# Can be used to find the owner of a data centre

import csv

# File locations (CCOD = domestic, OCOD = overseas)
CCOD_FILE = ""
OCOD_FILE = ""

def search_file(file_path, postcode_input, label):
    matches = 0
    with open(file_path, newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)

        for row in reader:
            # Normalise postcode + address
            row_postcode = (row.get("Postcode") or "").replace(" ", "").upper()
            row_address = (row.get("Property Address") or "").replace(" ", "").upper()
            row_address_nospace = row_address.replace(" ", "")

            if (row_postcode == postcode_input) or (postcode_input in row_address_nospace):
                matches += 1
                company = row.get("Proprietor Name (1)", "N/A")
                address = row.get("Property Address", "N/A")
                print(f"[{label}] {company} – {address} – {row.get('Postcode','')}")
    return matches

running = True

while running:
    postcode_input = input("Enter postcode to search: ").replace(" ", "").upper()

    total_matches = 0
    total_matches += search_file(CCOD_FILE, postcode_input, "CCOD")
    total_matches += search_file(OCOD_FILE, postcode_input, "OCOD")

    if total_matches == 0:
        print(f"No entries found for postcode {postcode_input}")
    else:
        print(f"Found {total_matches} entries for postcode {postcode_input}")

    retry = input("Would you like to run again? y/n: ")
    if retry != "y":
        running = False
```

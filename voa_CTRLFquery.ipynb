```python
import re
import csv

# SETUP
#input_file = ""
input_file = ""
output_file = ""

# KEYWORDS + VARIATIONS
keywords = [
    r"computer\s*centre",
    r"computer\s*center",
    r"data\s*centre",
    r"data\s*center",
    r"data\s*hall",
    r"\bserver\b", # avoids words like servery
]

# regex for case-insensitive search
pattern = re.compile("|".join(keywords), re.IGNORECASE)

# INIT VALUES
total_entries = 0
matches = []

# PROCESS FILE
with open(input_file, "r", encoding="utf-8") as infile:
    for line in infile:
        total_entries += 1
        normalized_line = re.sub(r"\s+", " ", line)  # normalise spaces
        if pattern.search(normalized_line):
            matches.append(line.strip())

# WRITE TO CSV
with open(output_file, "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    for entry in matches:
        writer.writerow([entry])  # each entry in its own row

# OUTPUT
print(f"Total entries scanned: {total_entries}")
print(f"Matches found: {len(matches)}")
print(f"Matched entries written to: {output_file}")
```
